```python
#factorize！

import numpy as np
import pandas as pd

data = np.random.choice(["a", "b", "c"], 10)
print(data)
# ['b' 'c' 'a' 'c' 'c' 'c' 'c' 'a' 'b' 'a']
labels, uniques = pd.factorize(data)
print(labels)
# [0 1 2 1 1 1 1 2 0 2]
print(uniques)
# ['b' 'c' 'a']
labels, uniques = pd.factorize(data, sort=True)
print(labels)
# [1 2 0 2 2 2 2 0 1 0]
print(uniques)
# ['a' 'b' 'c']
```

```python
# 分類モデルの入力データとしてカテゴリ変数を与える場合は適切なエンコードが必要
# カテゴリ変数を整数としてエンコードする
encoded_values, unique_values = pd.factorize(df["os_id"])#数据对应的编号，和不同的数据的个数
# クリックデータのカテゴリ変数を整数としてエンコードする
cat_cols = ['os_id', 'region']
for c in cat_cols:
    vv, obj = pd.factorize(df[c])
    df[c] = vv
# ここで紹介したのは大雑把なやり方
# モデルによっては他のエンコード方法のほうが適している場合もある
```
